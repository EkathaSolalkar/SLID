# Reading folder of wav files

```python
import glob
myfolder1="/content/drive/MyDrive/MiniProject/Dataset/Bengali"
myfolder1
```

```python
def getwave(path):
  # print(path)
  wave=glob.glob(path + "/*.wav")
  # print(wave)
  plt.figure(figsize=(14,5))
  for i in wave:
    signal,sample_rate=librosa.load(i)
    librosa.display.waveplot(signal,sr=1000)
```

```python
def preprocess(signal, samplerate):
    print('Sample rate: ', samplerate)
    sum=0
    for i in range(len(signal)):
        sum = sum+signal[i]
    print('sum : ', sum)
```

```python
def read_wav_file(path):
    wave=glob.glob(path + "/*.wav")
    for i in wave:
      print('wave : ', i)
      signal, sample_rate=librosa.load(i)
      print('length : ', len(signal))
      preprocess(signal, sample_rate)
      print('signal :', signal)

      print('--------------------------------------------------------------------------')
      #librosa.display.waveplot(signal,sr=1000)
      #print('---------')
```

```python
path = '/content/drive/MyDrive/MiniProject/Dataset/Bengali'
read_wav_file(path)
```
